```python
import requests
from bs4 import BeautifulSoup as soup
import pandas as pd
import csv
```

```python
#booking.com url
html = requests.get('https://www.booking.com/searchresults.html?ss=Banff%2C+Alberta%2C+Canada&label=gen173nr-1FCAEoggI46AdIM1gEaCeIAQGYATG4ARfIAQzYAQHoAQH4AQKIAgGoAgO4AvO0nK0GwAIB0gIkYzZmMjJkMmUtY2E0Yy00NmM5LWE2YzgtNjlmNjk1YmZkOGRk2AIF4AIB&aid=304142&lang=en-us&sb=1&src_elem=sb&src=index&dest_id=-560592&dest_type=city&ac_position=0&ac_click_type=b&ac_langcode=en&ac_suggestion_list_length=5&search_selected=true&search_pageview_id=308500f9986c001f&ac_meta=GhAzMDg1MDBmOTk4NmMwMDFmIAAoATICZW46BWJhbmZmQABKAFAA&group_adults=2&no_rooms=1&group_children=0&sb_travel_purpose=leisure')


```

```python
#initialising bsobj
bsobj=soup(html.content,'lxml')
```

```python
hotel = []
for name in bsobj.findAll('h3',{'class':'e1eebb6a1e e6208ee469 d0caee4251'}):
  hotel.append(name.text.strip())
```

```python
hotels=pd.DataFrame(hotel)
hotels
```

```python
review = []
for name in bsobj.findAll('div',{'class':'bui-review-score__title'}):
  review.append(name.text.strip())
```

```python
reviews=pd.DataFrame(review)
reviews
```

```python
rating=[]
for name in bsobj.findAll('div',{'class':'bui-review-score__badge'}):
  rating.append(name.text.strip())
```

```python
ratings=pd.DataFrame(rating)
ratings
```

```python

price_from=[]
for name in bsobj.findAll('div',{'class':'bui-price-display__value bui-f-color-constructive'}):
  price_from.append(name.text.strip())
```

```python
price=pd.DataFrame(price_from)
price
```

```python
frames=pd.concat([hotels, ratings,price, reviews], axis=1, join='inner')
frames.columns=['Hotels','Ratings','Price','Reviews']
frames
```
